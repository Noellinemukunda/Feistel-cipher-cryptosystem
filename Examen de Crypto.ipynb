# Algorithme de chiffrement

```python
a=int(input("Entrer un nombre de longueur 8 bits :"))
b=int(input("Entrer une permutation : "))
cle1=int(input("Entrer la première clé "))
cle2=int(input("Entrer la deuxième clé "))
tab=[]
mab=[]
perm=[]
cle11=[]
cle22=[]
m=0
n=0
r=0
y=0
w=0
for i in range(0,8,1):
    c=int(b/10**(7-i))
    d=c
    c=c-m
    tab.append(c)
    m=d*10
for j in range(0,8,1):
    f=int(a/10**(7-j))
    g=f
    f=f-n
    mab.append(f)
    n=g*10
print(tab)
print(mab)
for i in range(0,8,1):
    k=tab[i]
    s=mab[k]
    perm.append(s)
print(perm)
Go=perm[:4]
Do=perm[4:]
print(F"Go={Go}")
print(F"Do={Do}")



print('Premier round')
pe=2013
petab=[2,0,1,3]
pe_Go=[]
D1=[]
for i in range(0,4,1):
    l=petab[i]
    t=Go[l]
    pe_Go.append(t)
print(f"La permutation 2013 sur Go nous donne {pe_Go}")
    
    
for j in range(0,4,1):
    x=int(cle2/10**(3-j))
    e=x
    x=x-y
    cle22.append(x)
    y=e*10
    
for i in range(0,4,1):
    q=int(cle1/10**(3-i))
    v=q
    q=q - w
    cle11.append(q)
    w=v*10

print(f"La première clé est {cle11}")
print(f"La première clé est {cle22}")

for i in range(0,4,1):
    if pe_Go[i] != cle11[i]:
        t=1
    else:
        t=0
    D1.append(t)
print(f"D1 = {D1}")

chris=[]
for i in range(0,4,1):
    if Go[i]==1 or cle11[i]==1:
        tr=1
    else:
        tr=0
    chris.append(tr)
print(f"Go v K1 ={chris}")

G1=[]
for i in range(0,4,1):
    if Do[i]!=chris[i]:
        th=1
    else:
        th=0
    G1.append(th)
print(f"G1 = {G1}")    




pe_G1=[]
for i in range(0,4,1):
    l=petab[i]
    t=G1[l]
    pe_G1.append(t)
print(f"P(G1)={pe_G1}")

print("Deuxième round")
D2=[]
for i in range(0,4,1):
    if pe_G1[i] != cle22[i]:
        th=1
    else:
        th=0
    D2.append(th)
print(f"D2 = {D2}")
env=[]
for i in range(0,4,1):
    if G1[i]==1 or cle22[i]==1:
        tr=1
    else:
        tr=0
    env.append(tr)
print(f"G1 v K2 ={env}")

G2=[]
for i in range(0,4,1):
    if D1[i] != env[i]:
        th=1
    else:
        th=0
    G2.append(th)
print(f"G2= {G2}")

h=[]
h=G2+D2
print(f"La concaténation de G2 et D2 nous donne{C}")

from sympy.combinatorics import Permutation
p = Permutation(tab)
inv = p.__invert__()
print(f"L'inverse de la permutation est {inv.array_form}") # [1, 3, 2, 0]

z=[]
for i in range(0,8,1):
    k=inv.array_form[i]
    s=C[k]
    z.append(s)
print(z)

```

# Algorithme de la génération des clés pour Feistel

```python
a=int(input("Entrer un nombre de longueur 8 bits :"))
b=int(input("Entrer une permutation : "))
off=int(input("un décalage pour la première clé : "))
but=int(input("un décalage pour la deuxième clé : "))


tab=[]
mab=[]
perm=[]

m=0
n=0
r=0
y=0
w=0
for i in range(0,8,1):
    c=int(b/10**(7-i))
    d=c
    c=c-m
    tab.append(c)
    m=d*10
for j in range(0,8,1):
    f=int(a/10**(7-j))
    g=f
    f=f-n
    mab.append(f)
    n=g*10
print(tab)
print(mab)
for i in range(0,8,1):
    k=tab[i]
    s=mab[k]
    perm.append(s)
print(perm)
k_1=perm[:4]
k_2=perm[4:]
print(F"K_1={k_1}")
print(F"K_2={k_2}")




k1=[]

for i in range(0,4,1):
    if k_1[i] != k_2[i]:
        t=1
    else:
        t=0
    k1.append(t)
print(f"k1 = {k1}")


k2=[]


for i in range(0,4,1):
    if k_1[i]==1 and k_2[i]==1:
        tr=1
    else:
        tr=0
    k2.append(tr)
print(f"k2 ={k2}")


def decaler(list, offset):
    i = offset % len(list)
    return list[i:] + list[:i]


x= decaler(k1, off)
y= decaler(k2, but)

#Veuillez entrer un décalge négatif pour le décalage de droite

print(f"La première clé est {x}")
print(f"La deuxième clé est {y}")
```
